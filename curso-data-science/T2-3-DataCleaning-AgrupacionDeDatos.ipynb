# Agregacion de datos por categoria

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500       #Sera el tama√±o de nuestro dataset
gender_data = []
income_data = []

for i in range (0, n):
    gender_data.append(np.random.choice(gender))   #Se generaran elecciones random de las opciones que tiene la lista gender
    income_data.append(np.random.choice(income))   #Elecciones aleatorias de las opciones que tiene la lista income
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
#Para generar listas de datos numericos podemos hacer lo siguiente
height = 160 + 30 * np.random.randn(n)
#En este caso es de alturas, decimos que la media de alturas son 160cm, la desviacion estandar es de 30cm y n datos
#Recordando la formula 
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income": income
    }
)
```

```python
data.head()
```

# Agrupacion de datos

```python
grouped_gender = data.groupby("Gender")
#Crea 1 objeto que tendra 2 atributos, uno sera el nombre y el grupo
```

```python
grouped_gender.groups
#Nos da los numeros de fila que corresponden a Female y a Male
#Esto resulta util cuando se tienen muchos grupos, y se necesitan hacer operaciones separadas para cada uno de estos
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
#Para un solo grupo
```

```python
#Para hacer por mas de una agrupacion
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

# Transformacion de variables

```python
zscore = lambda x : (x - x.mean()) / x.std()
#A todos los elementos de la columna les restamos su promedio y los dividimos entre su desviacion estandar
```

```python
z_group = double_group.transform(zscore)
#Con esto tipificamos los datos
z_group
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group["Age"])
```

```python
#Podemos utilizar lambda para rellenar tambien valores faltantes dentro de nuestro dataset
fill_na_mean = lambda x : x.fillna(x.mean())
#Reemplazamos los datos que se encuentran en na con el promedio
```

```python
double_group.transform(fill_na_mean)
```
