# Carga de datos a traves de la funcion open

```python
data3 = open(mainpath + "/" + "customer-churn-model/Customer Churn Model.txt", "r")
#r para lectura, rw para lectura-escritura, w para escritura
#Open lo utilizamos para datasets enormes (20GB, por ejemplo)
#Nos ayuda a no emplear tanta memoria RAM
```

```python
cols = data3.readline().strip().split(",")
#next es para ir a la siguiente linea
#strip elimina los espacios en blanco que sobran al inicio y al final de la linea
#split crea un array a partir de divir todo por "," en este caso
#readline toma el valor solo de una linea
cols
```

```python
n_cols = len(cols)
#len(cols) es el numero de elementos dentro del array cols
n_cols
```

```python
counter = 0
main_dict = {}   #Es un diccionario vacio
for col in cols:
    main_dict[col] = []      #Cada una de las columnas sera la clave
```

```python
main_dict
```

```python
for line in data3:       #Iteramos en cada una de las lineas del dataset, guardado en data3
    values = line.strip().split(",")     #Se almacena cada valor del dataset
    for i in range(len(cols)):      #Iteramos en cada una de las columnas
        main_dict[cols[i]].append(values[i])      #A cada columna le agregamos el valor segun la fila en la que estemos
    counter += 1
print("El data set tiene %d filas y %d columnas" %(counter, n_cols))
#Ahora maindict tiene todo el dataset, ahora lo podemos convertir a un dataframe
```

```python
df3 = pd.DataFrame(main_dict)    #Transformamos el maindict, que contiene todo el dataset a un dataframe
df3.head()
```

# Ejercicio de descarga y procesamiento desde una URL

```python
import urllib3
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
r.status     #El status debe ser 200, eso significa que el estado es bueno
response = r.data
```

```python
str_data = response.decode('utf-8')
#Lo transformamos de un string binario a un string
```

```python
#Dividimos el string en un array de filas, separandolo por intros "\n"
lines = str_data.split("\n")
```

```python
#Recordemos que la primera linea es la cabecera, asi que la extraemos
col_names = lines[0].split(",")
n_cols = len(col_names)
```

```python
#Generamos un diccionario vacio donde ira la informacion procesada desde la URL externa
counter = 0
main_dict = {}
for col in col_names:
    main_dict[col] = []
```

```python
#Procesamos fila a fila la informacion para ir rellenando el diccionario con los datos como hicimos antes
for line in lines:
    #Nos saltamos la primera linea que es la que contiene la cabecera y ya tenemos procesada
    if (counter > 0):
        #Dividimos cada string por las comas como elemento separador
        values = line.strip().split(",")
        #Aniadimos cada valor a su respectiva columna del diccionario
        for i in range (len(col_names)):
            main_dict[col_names[i]].append(values[i])
    counter += 1
print("El dataset tiene %d filas y %d columnas" %(counter, n_cols))
```

```python
#Convertimos el diccionario procesado a Data Frame y comprobamos que los datos son correctos
medals_df = pd.DataFrame(main_dict)
medals_df.head()
```

```python
#Elegimos donde guardarlo (en la carpeta athletes es donde tiene mas sentido por el contexto del analisis)
mainpath = "/Users/Rudy/Documents/Github/python-ml-course/datasets"
filename = "athletes/winter_medals."
fullpath = os.path.join(mainpath, filename)
```

```python
#Lo guardamos en csv, json o en excel, segun queramos
medals_df.to_csv(fullpath + "csv")
medals_df.to_json(fullpath + "json")
medals_df.to_excel(fullpath + "xls")
```
